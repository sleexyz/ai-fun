```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters
max_len = 100  # Sequence length
d_model = 16  # Embedding dimension

position = np.arange(max_len).reshape(max_len, 1)
div_term = np.exp(np.arange(0, d_model, 2) * -(np.log(10000.0) / d_model))

# Positional Encodings
pe = np.zeros((1, max_len, d_model))
pe[0, :, 0::2] = np.sin(position * div_term)
pe[0, :, 1::2] = np.cos(position * div_term)

# Plot
plt.figure(figsize=(15, 10))
for i in range(d_model):
    if i == 1:
        plt.plot(pe[0, :, i], label=f'Dimension {i}')

plt.title('Positional Encoding Values')
plt.xlabel('Position')
plt.ylabel('Encoding value')
plt.legend()
plt.grid(True)
plt.show()
```
